```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_excel("OnlineRetail.xlsx")

print("Dataset Loaded Successfully")
print(data.head())

data.dropna(inplace=True)

data = data[~data['InvoiceNo'].astype(str).str.startswith('C')]

data = data[(data['Quantity'] > 0) & (data['UnitPrice'] > 0)]

print("\nData Cleaning Completed")
print("Dataset Shape:", data.shape)

user_item_matrix = data.pivot_table(
    index='CustomerID',
    columns='StockCode',
    values='Quantity',
    aggfunc='sum'
)

# Replace NaN with 0
user_item_matrix.fillna(0, inplace=True)

print("\nUser–Item Matrix Created")
print(user_item_matrix.head())

user_similarity = cosine_similarity(user_item_matrix)

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)

print("\nML Model Trained Successfully")

def recommend_products(customer_id, top_n=5):

    if customer_id not in user_similarity_df.index:
        return "Customer ID not found"

    similar_users = user_similarity_df.loc[customer_id] \
                                      .sort_values(ascending=False) \
                                      .iloc[1:6]

    product_scores = user_item_matrix.loc[similar_users.index].sum()

    purchased_products = user_item_matrix.loc[customer_id]
    product_scores = product_scores[purchased_products == 0]

    return product_scores.sort_values(ascending=False).head(top_n)

sample_customer_id = user_item_matrix.index.tolist()[0]

print("\nRecommended Products for Customer:", sample_customer_id)
recommendations = recommend_products(sample_customer_id)

print(recommendations)
```
